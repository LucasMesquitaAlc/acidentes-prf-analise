### Com a EDA feita, agora será feito testes estatísticos para maior rigor nas informações observadas nos dados

Para isso, usaremos a biblioteca Scipy


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
df = pd.read_csv('../dados/dados_tratados/prf_tratado.csv')
df.head()
```

O dia da semana influencia a fatalidade de um acidente?

Hipótese-nula: Todos os dias são igualmente fatais <br>
Hipótese alternativa: O dia da semana influencia a fatalidade

p-value < 5%

```python
#Criar tabela de contigência

tab_cont_dia = pd.crosstab(df['acidente_fatal'],df['dia_semana'])
tab_cont_dia
```

```python
#MEDIR CHI²

chi2, p, gl, esperado = scipy.stats.chi2_contingency(tab_cont_dia)

print(f'χ² = {chi2:.2f}')
print(f'Valor-p = {p:.4e}')
print(f'Graus de liberdade = {gl}')
```

```python
#DESCOBRIR O V de Cramér

n = tab_cont_dia.sum().sum()
v_cramer = np.sqrt(chi2 / (n * (min(tab_cont_dia.shape) - 1)))
print(f"V de Cramér = {v_cramer:.3f}")
```

Com o valor-p de 2.78×10⁻⁷², podemos, com confiança, negar a hipótese nula e concluir que o dia da semana afeta a taxa de fatalidade.

Porém, com o V de Cramér em 0.033, a associação é considerada de magnitude fraca, algo que é reforçado com a taxa de fatalidade por dia da semana visto anteriormente
