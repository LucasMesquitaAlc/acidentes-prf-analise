```python
import pandas as pd
import numpy as np
```

```python
df2021 = pd.read_csv('../dados/dados_brutos/datatran2021.csv',sep=';', encoding='latin-1')
df2022 = pd.read_csv('../dados/dados_brutos/datatran2022.csv',sep=';', encoding='latin-1')
df2023 = pd.read_csv('../dados/dados_brutos/datatran2023.csv',sep=';', encoding='latin-1')
df2024 = pd.read_csv('../dados/dados_brutos/datatran2024.csv',sep=';', encoding='latin-1')
df2025 = pd.read_csv('../dados/dados_brutos/datatran2025.csv',sep=';', encoding='latin-1')

df = pd.concat([df2021,df2022, df2023, df2024, df2025],ignore_index=True)
df
```

```python
#As colunas "regional", "delegacia" e "uop" são desnecessárias, então removemos do df

df = df.drop(columns=['regional', 'delegacia', 'uop'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['data_inversa'] = pd.to_datetime(df['data_inversa'])
```

```python
df.isnull().sum()
#5 valores nulos apenas
```

```python
df['classificacao_acidente'].value_counts()
```

```python
#Lidando com os valores nulos na coluna 'classificacao_acidente'
condicoes = [
    df['mortos'] > 0,
    df['feridos'] > 0,
    df['feridos'] == 0
]

classificacao = ['Com Vítimas Fatais', 'Com Vítimas Feridas', 'Sem Vítimas']
df['classificacao_acidente'] = np.select(condicoes, classificacao)
```

```python
df['classificacao_acidente'].value_counts()
```

```python
#Criar coluna de feriado

import holidays
feriados_br = holidays.Brazil(years=[2021, 2022, 2023, 2024, 2025])
df['feriado'] = df['data_inversa'].isin(feriados_br)

df.head(15)
```

```python
#Criar coluna "acidente_fatal" (variável-alvo para futuro projeto de ML)

df['acidente_fatal'] = np.where(df['mortos'] > 0, 1,0)
df['acidente_fatal'].value_counts()
```

```python
df['latitude'] = df['latitude'].str.replace(',', '.').astype(float)
df['longitude'] = df['longitude'].str.replace(',', '.').astype(float)
```

```python
df.to_csv('../dados/dados_tratados/prf_tratado.csv', index=False)
```
